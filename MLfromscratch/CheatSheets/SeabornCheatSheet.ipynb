```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")  # Built-in dataset in Seaborn
print(df.head())
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=df)
plt.show()
```

```python
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.kdeplot(df['total_bill'], fill=True)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', split=True)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df, ci='sd')
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.show()
```

```python
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
sns.pairplot(df, hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
g = sns.FacetGrid(df, col='sex')
g.map(sns.histplot, 'total_bill')
plt.show()
```

### Customizing Seaborn Styles

```python
sns.set_style("darkgrid")  # Other styles: whitegrid, dark, white, ticks
sns.set_palette("husl")  # Color palettes: deep, muted, bright, colorblind
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

### Adding Titles and Labels

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.title("Total Bill vs Tip Relationship")
plt.show()
```
