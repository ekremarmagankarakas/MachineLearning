```python
import matplotlib.pyplot as plt
```

### Basic Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)
plt.show()
```

```python
### Customizing Line Style and Color
```

```python
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='black')
plt.show()
```

### Multiple Lines in One Plot

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]
y3 = [15, 25, 35, 45, 55]
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--')
plt.plot(x, y3, label='Line 3', color='purple', linestyle=':')
plt.legend()
plt.show()
```

### Scatterplot

```python
plt.plot(x, y, 'o', color='purple', markersize=10, alpha=0.7)  # 'o' makes it a scatter plot
plt.show()
```

### Adding Labels, Title, and Legend

```python
plt.plot(x, y, label='Growth')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Line Plot Example")
plt.legend()
plt.show()
```

### Adding Axes and Grid

```python
plt.plot(x, y)
plt.xlim(0, 6)  # Set X-axis range
plt.ylim(5, 45)  # Set Y-axis range
plt.xticks([0, 1, 2, 3, 4, 5, 6])  # Custom X-axis ticks
plt.yticks([5, 10, 20, 30, 40, 50])  # Custom Y-axis ticks
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

### Logarithmic Scale

```python
plt.plot(x, y)
plt.yscale('log')
plt.xscale('log')
plt.show()
```

### Subplots

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0, 0].plot(x, y, color='blue')
ax[0, 1].plot(x, y, color='red')
ax[1, 0].plot(x, y, color='green')
ax[1, 1].plot(x, y, color='orange')
plt.show()
```

### Filling Area Under a Curve

```python
plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='lightblue', alpha=0.4)
plt.show()
```

### Adding Annotations

```python
plt.plot(x, y)
plt.annotate("Peak", xy=(5, 40), xytext=(3, 35),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

### Plotting with Different Styles

```python
plt.style.use('ggplot')  # Other options: 'seaborn', 'classic', 'bmh', 'dark_background'
plt.plot(x, y)
plt.show()
```

### Twin Axes (Dual Y-Axis)

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y1', color='g')
ax2.set_ylabel('Y2', color='b')
plt.show()
```

### Histogram

```python
import numpy as np
np.random.seed(42)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

### Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 8]
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

### Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'])
plt.title("Pie Chart Example")
plt.show()
```

### Error Bars

```python
errors = [1, 2, 1, 1, 2]
plt.errorbar(x, y, yerr=errors, fmt='-o', capsize=5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Error Bar Plot")
plt.show()
```

### Plotting from pandas DF

```python
import pandas as pd

# Creating a sample DataFrame
np.random.seed(42)
df = pd.DataFrame({
    'X': np.arange(1, 11),
    'Y1': np.random.randint(10, 100, 10),
    'Y2': np.random.randint(5, 50, 10),
    'Category': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D', 'A', 'B']
})

# Line Plot
plt.figure(figsize=(6, 4))
plt.plot(df['X'], df['Y1'], marker='o', linestyle='-', color='b', label='Y1')
plt.plot(df['X'], df['Y2'], marker='s', linestyle='--', color='r', label='Y2')
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.title("Line Plot")
plt.legend()
plt.grid(True)
plt.show()

# Scatter Plot
plt.figure(figsize=(6, 4))
plt.scatter(df['X'], df['Y1'], color='g', label='Y1')
plt.scatter(df['X'], df['Y2'], color='m', label='Y2')
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.title("Scatter Plot")
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(df['X'], df['Y1'], color='c', label='Y1', alpha=0.7)
plt.bar(df['X'], df['Y2'], color='orange', label='Y2', alpha=0.7)
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.legend()
plt.show()

# Histogram
plt.figure(figsize=(6, 4))
plt.hist(df['Y1'], bins=5, color='b', alpha=0.7, label='Y1')
plt.hist(df['Y2'], bins=5, color='r', alpha=0.7, label='Y2')
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.legend()
plt.show()

# Pie Chart
plt.figure(figsize=(6, 6))
category_counts = df['Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'])
plt.title("Pie Chart of Categories")
plt.show()

# Error Bar Plot
errors = np.random.randint(1, 10, size=len(df['X']))
plt.figure(figsize=(6, 4))
plt.errorbar(df['X'], df['Y1'], yerr=errors, fmt='-o', capsize=5, label='Y1')
plt.errorbar(df['X'], df['Y2'], yerr=errors, fmt='--s', capsize=5, label='Y2')
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.title("Error Bar Plot")
plt.legend()
plt.show()
```
